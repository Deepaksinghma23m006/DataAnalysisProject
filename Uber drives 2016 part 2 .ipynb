# grouping 

```python
#read the data
import pandas as pd
df=pd.read_csv("Uber Drives 2016.csv")
```

```python
df.head()
```

```python
# city wise means of miles
df.groupby("START*")["MILES*"].agg("mean")  # we can use min,max,sd etc. for more fn. we will use numpy
```

```python
#or
df.groupby("START*")["MILES*"].mean()
```

```python
#or
df.groupby("START*").mean()      # if there is only one float column
```

Q for each start &stop location find mean& sum resp.the total distance travelled

```python
df.groupby(["START*","STOP*"]).agg(["mean","sum"]).head(20)
```

Q find most recent &earliest travelled date & mean distance travelled for each start city

```python
df.groupby("START*").agg({"START_DATE*":["min","max"],"MILES*":["mean"]})  #my way
```

```python
# by sir , we have to do firstly covert date to datetime format
# firstly remove last row because its containing no value (do check by df.tail())
df.iloc[:-1,:]
df["START_DATE*"]=pd.to_datetime(df["START_DATE*"])
df.dtypes
```

```python
res=df.groupby("START*").agg({"START_DATE*":["min","max"],"MILES*":["mean"]})
res
```

```python
# now to reset index and change column headings
res.reset_index(inplace=True)
res
```

```python
res.columns=["start_city","recent trp","farthest_trip","average distance"] #to change column headings
res
```

# to save filein csv format

```python
res.to_csv("raghu.csv")  # file name = raghu
```
